# Early stages of cleaning the data

```python
zhvi=zhvi.rename(columns={'RegionName':'ZIP'})
FLzhvinona = zhvi[zhvi["State"]=='FL']
FLzhvi=FLzhvinona.dropna()
FLzhvi
```

# 1996-2017 overall Average Home Value

```python
overallavg=FLzhvi.groupby(['SizeRank','ZIP','State','City','CountyName']).mean()
overallavg
overallavg.transform('mean',axis=1)
```

# Yearly 12 month Average Home Value for individual years 1997-2001

# 1997

```python
ZCS1997 = FLzhvi[['SizeRank','ZIP','City','State','1997-01','1997-02','1997-03','1997-04','1997-05','1997-06',
                  '1997-07','1997-08','1997-09','1997-10','1997-11','1997-12']].copy()
Avg1997=ZCS1997.groupby(['SizeRank','ZIP','City','State']).mean()
T1997=Avg1997.transform('mean', axis=1)
T1997
```
