### Import Libraries

```python
%pylab inline
import numpy, csv
import pandas as pd
import matplotlib.pyplot as plt
```

## Load CSV Files
The goal of this notebook is to explore some TRI data for a 5 year period (2010-2014) in the state of Michigan. I chose this state because it has a relatively dense amount of industrial production facilities.

```python
mi_10 = pd.read_csv('./data/TRI_Files/TRI_2010_MI.csv')
mi_11 = pd.read_csv('./data/TRI_Files/TRI_2011_MI.csv')
mi_12 = pd.read_csv('./data/TRI_Files/TRI_2012_MI.csv')
mi_13 = pd.read_csv('./data/TRI_Files/TRI_2013_MI.csv')
mi_14 = pd.read_csv('./data/TRI_Files/TRI_2014_MI.csv')

data_yrs = [mi_10,mi_11,mi_12,mi_13,mi_14]
```

### Extracting Relevant Columns
For this exercise, I'll be interested in the facility names, chemicals reported, and total amount of each chemical released.

```python
nameChemTotal = [yr[['YEAR','FACILITY_NAME','CHEMICAL','TOTAL_RELEASES']] for yr in data_yrs]
```

#### Top 10 For Each Year
Here, I extract the top 10 facilities for each imported year in terms of their overall releases

```python
topTens = [nameChemTotal[i].sort_values('TOTAL_RELEASES', ascending=False)[0:10] for i in range(len(nameChemTotal))]
```

## List and Plot Top 10 Facilities Releasing Carcinogens for Each Year
***

```python
carcReleases = [yr[['YEAR','FACILITY_NAME','CHEMICAL','CARCINOGEN','TOTAL_RELEASES']] for yr in data_yrs]
```
