## Start of basemap
Here I will attempt to plot TRI reporting facilities with on-site carcinogen based releases

```python

%pylab inline
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy, csv
import pandas as pd

NC2002 = pd.read_csv("data\TRI_2002_NC.csv")
NC2003 = pd.read_csv("data\TRI_2003_NC.csv")
NC2004 = pd.read_csv("data\TRI_2004_NC.csv")
NC2005 = pd.read_csv("data\TRI_2005_NC.csv")
NC2006 = pd.read_csv("data\TRI_2006_NC.csv")

NC_data =[NC2002,NC2003,NC2004,NC2005,NC2006]
```

```python
nc = Basemap(projection='merc', resolution='h',
                     llcrnrlon = -84.5, urcrnrlon = -75.2,
                     llcrnrlat = 34.0, urcrnrlat = 36.9)
```

```python
NCthis=NC2006[NC2006['CARCINOGEN']== 'YES']
someonsite=NCthis[NCthis['ON-SITE_RELEASE_TOTAL']>0]
final= someonsite[['FACILITY_NAME','CITY','YEAR','LATITUDE','LONGITUDE','FACILITY_NAME','ON-SITE_RELEASE_TOTAL']]
lon06, lat06=nc(final.LONGITUDE.values,final.LATITUDE.values)
```

## Cool stuff goes here 

```python

plt.figure(figsize = (25,25))
nc.drawcountries()
nc.drawstates(linewidth = 3)
nc.drawcoastlines()
nc.drawcounties(linewidth = .5, color='white')
nc.fillcontinents(color = 'powderblue')
nc.drawmapboundary()
nc.scatter(lon06, lat06,c='black', marker='o', alpha=1, zorder=10, cmap='black')
plt.title("TRI facilities with On-Site Releases 2006")
```
